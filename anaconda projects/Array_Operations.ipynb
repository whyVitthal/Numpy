Broadcasting :
Broadcasting in NumPy allows us to perform arithmetic operations on arrays of different shapes without reshaping them. It automatically adjusts the smaller array to match the larger array's shape by replicating its values along the necessary dimensions. This makes element-wise operations more efficient by reducing memory usage and eliminating the need for loops.

```python
l = [1,2,3,4]
arr=np.array(l)
arr+10
# agar ek sinlge element ko sabme add krna chah rhe ho tb broadcasting us ekr skte h
```

```python
arr2 = np.arange(1,26).reshape(5,5)
arr2
```

```python
arr2 +10
```

```python
# Deep and Shallow copy
```

Deep Copy: Creates a completely independent copy of the object and all nested objects.
Shallow Copy: Creates a new outer object, but nested objects are shared with the original.

```python
a = np.arange(1,21)
a
```

```python
slice = a[:5]
slice = slice+10
slice
```

```python
a
```

```python
# b = a
b = a.copy()
b[0]=99
b
```
